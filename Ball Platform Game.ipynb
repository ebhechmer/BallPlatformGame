```python
import pygame
from brian2 import *

# starting pygame
pygame.init()

# setting screen size
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Ball and Platform Game")

# Create a clock for managing framerate
clock = pygame.time.Clock()

# setting initial platform that ball will bounce off of
platform_pos = [300, 500]
platform_size = (200,20)
#---------------------------
platform_angle = 0
platform_surf = pygame.Surface(platform_size, pygame.SRCALPHA)
platform_surf.fill((0, 0, 0))
#---------------------------

# setting a ball initial position and radius
ball_pos = [400, 300]
radius = 20

# setting initial velocity of the ball to 0
ball_vel = [0, 0]

# setting up gravity
gravity = 0.1

# Define the model
eqs = '''
dtheta/dt = k*(x - cx) : radian
'''

# Create the network
net = Network()
net.add(NeuronGroup(1, eqs, method="euler"))

# Set the initial conditions
net.theta = 0
net.k = 1*Hz
net.cx = 200

# Set the input
net.x = ball_pos[0]

# Define the update function
def update_platform_angle():
    net.x = ball_pos[0]
    net.theta = net.theta + net.k * (net.x - net.cx)
    net.cx = net.x
    platform_angle = net.theta
    rotated_platform_surf = pygame.transform.rotate(platform_surf, platform_angle)
    platform_rect = rotated_platform_surf.get_rect()
    #platform_rect.center = platform_pos
    platform_rect.center = (net.x, 500)
    screen.blit(rotated_platform_surf, platform_rect)

# def change_velocity():
#     if net.theta < 90:
#         if ball_vel[0] > 0:
#             ball_vel[0] = -ball_vel[0]
#         else:
#             ball_vel[0] = ball_vel[0]
#     elif net.theta[0] > 90 and net.theta[0] < 180:
#         if ball_vel[0] > 0:
#             ball_vel[0] = -ball_vel[0]
#     else:
#         ball_vel[0] = -ball_vel[0]

#-----------------------------
running = True
while running:
    # keeps framerate at 60
    clock.tick(60)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                ball_vel[0] = -2
            elif event.key == pygame.K_RIGHT:
                ball_vel[0] = 2
            elif event.key == pygame.K_UP:
                ball_vel[1] = -2
            elif event.key == pygame.K_DOWN:
                ball_vel[1] = 2
            elif event.key == pygame.K_q:
                print("Game has been quit.")
                running = False
#         elif event.type == pygame.KEYUP:
#             if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
#                 ball_vel[0] = 0
#             elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
#                 ball_vel[1] = 0
                
    #updating platform angle
    #platform_angle += 2
    
    # Apply gravity to ball velocity
    ball_vel[1] += gravity
    
    # Update ball position based on velocity
    ball_pos[0] += ball_vel[0]
    ball_pos[1] += ball_vel[1]

    # Check for collision with platform
#     if ball_pos[1] + radius >= platform_pos[1]:
#         if (ball_pos[0] + radius > platform_pos[0] and
#             ball_pos[0] - radius < platform_pos[0] + platform_size[0]):
#             # Reverse ball velocity on y-axis
#             ball_vel[1] = -ball_vel[1]
#             pygame.time.wait(10)
    
    if ball_pos[1] + radius >= 500:
        if ball_pos[1] + radius >= ball_pos[0]:
            ball_vel[1] = -ball_vel[1]
            #change_velocity()
            ball_vel[0] = -ball_vel[0]
            pygame.time.wait(10)

    # Clear screen
    screen.fill((255, 255, 255))
    
    update_platform_angle()
    #-------------------------------------------------------------
    # Create a rotated copy of the platform surface
    #rotated_platform_surf = pygame.transform.rotate(platform_surf, platform_angle)
    # grab position of cursor
    #mx, my = pygame.mouse.get_pos()
    # Draw rotated platform
    #screen.blit(rotated_platform_surf, (mx - int(rotated_platform_surf.get_width())/2, my - int(rotated_platform_surf.get_height())/2))
    #screen.blit(platform_surf, (300,400))
    #-------------------------------------------------------------
    # drawing the ball and platform
    pygame.draw.circle(screen, (255, 0, 0), ball_pos, radius)
    #pygame.draw.rect(screen, (0, 0, 0), (platform_pos[0], platform_pos[1], platform_size[0], platform_size[1]))

    # checking if the ball is off screen
    if (ball_pos[0] - radius < 0 or 
        ball_pos[0] + radius > screen.get_width() or
        ball_pos[1] - radius < 0 or
        ball_pos[1] + radius > screen.get_height()):
        print("The ball has went off screen.")
        running = False

    # updating game display every loop
    pygame.display.update()

pygame.quit()
```
